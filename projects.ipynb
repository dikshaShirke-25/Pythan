```python
sub1= int(input("enter 1st subject:"))
sub2= int(input("enter 2nd subject:"))
sub3= int(input("enter 3rd subject:"))
sub4= int(input("enter 4th subject:"))
sub5= int(input("enter 5th subject:"))
per=(sub1+sub2+sub3+sub4+sub5)/500*100
print(per,"%")
if per<0:
    print("AUR PHONE CHALAO")
elif per<35:
    print("fail")
elif per<40:
    print("pass")
elif per<60:
    print("second class")
elif per<75:
    print("first class")
elif per>101:
    print("not applicable")
else:
    print("distinction")
```
